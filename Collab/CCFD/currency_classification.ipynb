```python

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
import cv2
import glob
import random
import tensorflow as tf
import tensorflow_hub as hub
import shutil
```

```python
import os

dataset_path = "/content/drive/MyDrive/FYP/CCFD/data/"
groundtruth_images = []
image_size = 512
patch_size = 64

subfolders = sorted([f for f in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, f))])

print(subfolders)
```

```python
subfolders_path = []

for subfolder in subfolders:
    subfolder_path = os.path.join(dataset_path, subfolder)
    subfolders_path.append(subfolder_path)

print(subfolders_path)
```

```python
labels = subfolders

images = []
image_labels = []

for label in labels:
  image_files = os.listdir(os.path.join(dataset_path, label))

  for image_file in image_files:

        image_path = os.path.join(dataset_path, label, image_file)

        img = cv2.imread(image_path)
        img = cv2.resize(img, (224, 224))

        images.append(img)
        image_labels.append(label)




```

```python

print(image_labels)

labels = []

for label in image_labels:

    if label == subfolders[0]:
        labels.append(0)

    elif label == subfolders[1]:
        labels.append(0)

    elif label == subfolders[2]:
        labels.append(1)

    elif label == subfolders[3]:
        labels.append(1)

    elif label == subfolders[4]:
        labels.append(2)

    elif label == subfolders[5]:
        labels.append(2)

    elif label == subfolders[6]:
        labels.append(3)

    elif label == subfolders[7]:
        labels.append(3)

    elif label == subfolders[8]:
        labels.append(4)

    elif label == subfolders[9]:
        labels.append(4)

    elif label == subfolders[10]:
        labels.append(5)

    elif label == subfolders[11]:
        labels.append(5)

    elif label == subfolders[12]:
        labels.append(6)

    elif label == subfolders[13]:
        labels.append(6)
```

```python
values, counts = np.unique(labels, return_counts=True)
print(values)
print(counts)
```

```python
X = np.array(images)
Y = np.array(labels)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.1,
                                                    random_state=7)

x_train = np.zeros((1000, 224, 224, 3))
y_train = np.zeros((1000))


x_train = X_train[:2500]
y_train = Y_train[:2500]


```

```python
X_train_scaled = x_train/255

X_test_scaled = X_test/255
```

```python
mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)
```

```python
num_of_classes = 7

model = tf.keras.Sequential([

    pretrained_model,
    tf.keras.layers.Dense(num_of_classes)

])

model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(X_train_scaled, y_train, epochs=20)
```

```python

score, acc = model.evaluate(X_test_scaled, Y_test)
#print('Test Loss =', score)
print('Test Accuracy =', acc*100)
```

```python
def classifyImage(path):

    input_image = cv2.imread(path)
    #plt.imshow(input_image)
    #plt.show()

    input_image_resize = cv2.resize(input_image, (224,224))

    input_image_scaled = input_image_resize/255

    image_reshaped = np.reshape(input_image_scaled, [1,224,224,3])

    input_prediction = model.predict(image_reshaped)

    print(input_prediction)

    input_pred_label = np.argmax(input_prediction)

    print(f"Label Predicted: {input_pred_label}")

    if(input_pred_label == 0):
      print("Rs 1000")
    if(input_pred_label == 1):
      print("Rs 100")
    if(input_pred_label == 2):
      print("Rs 10")
    if(input_pred_label == 3):
      print("Rs 20")
    if(input_pred_label == 4):
      print("Rs 5000")
    if(input_pred_label == 5):
      print("Rs 500")
    if(input_pred_label == 6):
      print("Rs 50")
```
